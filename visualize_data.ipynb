```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv", infer_datetime_format=True, parse_dates=True)
df
```

```python
df.magnitude = df.magnitude.round(1)
```

```python
df.date = pd.to_datetime(df.date)
```

```python
df = df[df['date'] < '2023-01-01']
```

```python
px.histogram(df, x="magnitude", histnorm='probability density')
```

```python
df2 = df.groupby(df['date'].map(lambda x: x.year))['eventID'].count().reset_index(name ='Count')
df2.date = df2.date.astype(str)
```

```python
px.histogram(df2, x="date", y="Count").update_layout(yaxis_title="Count")
```
